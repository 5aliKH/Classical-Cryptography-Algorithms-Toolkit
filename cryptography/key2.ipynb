```python
 #خوارزميه قيصر
def caesar_cipher(text, shift):
    result = ""  # إنشاء سلسلة فارغة لتخزين النص المشفر
    for char in text:  # المرور على كل حرف في النص
        if char.isalpha():  # التحقق إذا كان الحرف حرفًا (ليس رقم أو رمز)
            # تحديد بداية الحرف (A أو a) بناءً على حالة الحرف
            start = ord('A') if char.isupper() else ord('a')
            # حساب الحرف المشفر باستخدام الصيغة
            # (حرف ASCII + الإزاحة) % 26 
            
            result += chr((ord(char) - start + shift) % 26 + start)
        else:
            result += char  # إضافة الحرف كما هو إذا لم يكن حرفًا
    return result  # إرجاع النص المشفر

def caesar_decipher(text, shift):
    return caesar_cipher(text, -shift)  # لفك التشفير، نستخدم الإزاحة السلبية

# الحصول على نص المستخدم ومفتاح الإزاحة
text = input("أدخل النص المراد تشفيره: ")
shift = int(input("أدخل رقم المفتاح (الإزاحة): "))

# تشفير النص
encrypted_text = caesar_cipher(text, shift)
print("النص المشفر:", encrypted_text)

# فك تشفير النص
decrypted_text = caesar_decipher(encrypted_text, shift)
print("النص المفكوك:", decrypted_text)
```

```python
#ATHASH CIPHER
def atbash_cipher(text):
    result = ""  # إنشاء سلسلة فارغة لتخزين النص المشفر
    for char in text:  # المرور على كل حرف في النص
        if char.isalpha():  # التحقق إذا كان الحرف حرفًا (ليس رقم أو رمز)
            # حساب الحرف المشفر باستخدام الصيغة Atbash
            # 'A' -> 'Z', 'B' -> 'Y', ... , 'a' -> 'z', 'b' -> 'y'
            if char.isupper():
                result += chr(ord('Z') - (ord(char) - ord('A')))
            else:
                result += chr(ord('z') - (ord(char) - ord('a')))
        else:
            result += char  # إضافة الحرف كما هو إذا لم يكن حرفًا
    return result  # إرجاع النص المشفر

# الحصول على نص المستخدم
text = input("أدخل النص المراد تشفيره: ")

# تشفير النص
encrypted_text = atbash_cipher(text)
print("النص المشفر:", encrypted_text)

# فك تشفير النص (Atbash هو متماثل، لذا نستخدم نفس الدالة)
decrypted_text = atbash_cipher(encrypted_text)
print("النص المفكوك:", decrypted_text)
```

```python
#خوارزميه hill
#مكتبه للعمليات العدديه وللتعامل مع المصفوفات
import numpy as np

# تحويل الحروف إلى أرقام والعكس (A=0, B=1, ..., Z=25)
def char_to_num(c):
    return ord(c) - ord('A')

def num_to_char(n):
    return chr(n + ord('A'))

# دالة لتحويل كلمة المفتاح إلى مصفوفة 2x2
def word_to_key_matrix(word):
    word = word.upper()
    if len(word) != 4:
        raise ValueError("يجب أن تكون كلمة المفتاح مكونة من 4 أحرف.")
    
    key = [char_to_num(c) for c in word]
    key_matrix = np.array([[key[0], key[1]], [key[2], key[3]]])
    return key_matrix

# دالة لحساب معكوس المصفوفة في الحقل Z26
def mod_inverse_matrix(matrix, mod=26):
    det = int(np.round(np.linalg.det(matrix)))  # حساب محدد المصفوفة
    det_inv = pow(det, -1, mod)  # حساب معكوس المحدد
    matrix_mod_inv = (det_inv * np.round(det * np.linalg.inv(matrix)).astype(int)) % mod
    return matrix_mod_inv

# تشفير النص باستخدام خوارزمية Hill
def hill_cipher(text, key_matrix):
    block_size = key_matrix.shape[0]
    blocks = [text[i:i+block_size] for i in range(0, len(text), block_size)]
    
    cipher_text = ""
    for block in blocks:
        block_vector = np.array([char_to_num(c) for c in block])
        encrypted_vector = np.dot(key_matrix, block_vector) % 26
        cipher_text += ''.join(num_to_char(int(c)) for c in encrypted_vector)
    return cipher_text

# فك التشفير باستخدام خوارزمية Hill
def hill_decipher(cipher_text, key_matrix):
    inv_key_matrix = mod_inverse_matrix(key_matrix)
    
    block_size = key_matrix.shape[0]
    blocks = [cipher_text[i:i+block_size] for i in range(0, len(cipher_text), block_size)]
    
    plain_text = ""
    for block in blocks:
        block_vector = np.array([char_to_num(c) for c in block])
        decrypted_vector = np.dot(inv_key_matrix, block_vector) % 26
        plain_text += ''.join(num_to_char(int(c)) for c in decrypted_vector)
    return plain_text

# عرض الخيارات للمستخدم لاختيار كلمة مفتاح
def get_key_from_word():
    keys = {
        "hill": "HILL",
        "zirp": "ZIRP",
        "urll": "URLL",  # أمثلة أخرى
        "test": "TEST",
        "neop": "NEOP"
    }
    
    print("اختر كلمة مفتاح من بين الخيارات التالية:")
    for i, (key_name, key_word) in enumerate(keys.items(), 1):
        print(f"{i}. {key_name} ({key_word})")
    
    choice = int(input("أدخل رقم الاختيار: "))
    key_word = list(keys.values())[choice - 1]
    
    return word_to_key_matrix(key_word)

# الحصول على النص من المستخدم (يجب أن يكون 4 أحرف فقط)
text = input("أدخل النص المراد تشفيره (4 أحرف فقط): ").upper()

if len(text) != 4:
    print("يجب إدخال 4 أحرف بالضبط.")
else:
    key_matrix = get_key_from_word()

    # التأكد من أن المصفوفة لها محدد غير صفري
    if int(np.round(np.linalg.det(key_matrix))) % 26 == 0:
        print("المفتاح غير صالح. محدد المصفوفة يجب ألا يكون صفرًا.")
    else:
        # تشفير النص
        encrypted_text = hill_cipher(text, key_matrix)
        print("النص المشفر:", encrypted_text)

        # فك التشفير
        decrypted_text = hill_decipher(encrypted_text, key_matrix)
        print("النص المفكوك:", decrypted_text)
```

```python
#خوارزميه atue key
# تحويل الحروف إلى أرقام والعكس (A=0, B=1, ..., Z=25)
def char_to_num(c):
    return ord(c) - ord('A')
#char_to_num  upper داله تعتمد على الاحرف الكبيره لذلك نستخدم
def num_to_char(n):
    return chr(n + ord('A'))

# تمديد المفتاح باستخدام النص الأصلي
def extend_key(key, text):
    extended_key = key + text[:len(text) - len(key)]
    return extended_key.upper()
#تحول الحروف الصغيره الى حروف كبيره
# تشفير النص باستخدام خوارزمية Auto-Key
def auto_key_cipher(text, key):
    text = text.upper()
    key = extend_key(key, text)
    
    cipher_text = ""
    for i in range(len(text)):
     
    # حساب الحرف المشفر (إضافة القيمة العددية للحرف والنص معاً)
        encrypted_char = (char_to_num(text[i]) + char_to_num(key[i])) % 26
        cipher_text += num_to_char(encrypted_char)
    
    #(char_to_num(text[i]) داله تحول الحرف الحالي من النص الى قيمته العدديه
    #(char_to_num(text[i]) + char_to_num(key[i])) جمع القيمه العدديه للحرفمع قيمه العدديه للمفتاح
    #num_to_char نقوم بتحويل القيمه العدديه الى حرف
    #نقوم بادخال النص المشفر الى داله cipher_text  
    return cipher_text

# فك التشفير باستخدام خوارزمية Auto-Key
def auto_key_decipher(cipher_text, key):
    cipher_text = cipher_text.upper()
    key = key.upper()
    #انشاء سلسله فارغه  لتخزين النص المفكوك 
    plain_text = ""
    #  for i in range(len(cipher_text)): المرور على كل حرف في النص المشفر
    for i in range(len(cipher_text)):
        # حساب الحرف المفكوك (طرح قيمة الحرف المشفر من المفتاح)
        decrypted_char = (char_to_num(cipher_text[i]) - char_to_num(key[i])) % 26
        plain_text += num_to_char(decrypted_char)
        # plain_text += num_to_char(decrypted_char)    هنا نحول الناتج الى حرف ونضيفه الى السلسه
        # تمديد المفتاح باستخدام الحروف المفكوكة
        key += plain_text[-1]
    #تمديد المفتاح عن طريق اضافه الحرف المفكوك الاخير
    return plain_text

# الحصول على النص والمفتاح من المستخدم
text = input("أدخل النص المراد تشفيره: ").upper()
key = input("أدخل المفتاح: ").upper()

# تشفير النص
encrypted_text = auto_key_cipher(text, key)
print("النص المشفر:", encrypted_text)

# فك تشفير النص
decrypted_text = auto_key_decipher(encrypted_text, key)
print("النص المفكوك:", decrypted_text)
```

```python
#خوارزميه لعبه الحنش
import math

# دالة لتشفير النص باستخدام Columnar Transposition Cipher
def encrypt_message(plaintext, key):
    # حساب عدد الأعمدة (طول المفتاح)
    num_columns = key
    
    # حساب عدد الصفوف عن طريق تقسيم طول النص على عدد الأعمدة
    #نحسب عدد الصفوف من خلال قسمه طول النص على عدد الاعمده
    #math.ceil تجعل النتيجه عددا صحيحا كاملا
    num_rows = math.ceil(len(plaintext) / num_columns)
    
    # num_columns عدد الاعمده
    # ملء النص في شبكة ذات أبعاد (num_rows, num_columns)
    matrix = [''] * num_columns
    #هذا النص يعني انه يضيف قايمه فارغه لكل عمود(العمود يعتمد على المفتاح الذي يدخله المستخدم)
    
    # إدخال النص في الأعمدة
    for index, char in enumerate(plaintext):
        column = index % num_columns
        matrix[column] += char
        #index العدد التسلسلي للحرف
    #char هو الحرف نفسه
    #index % num_columns يساعد في توزيع الحروف على الاعمده بطريقه دائريه
    # matrix[column] += char    اضافه الحرف الى العمود المناسب في المصفوفه  
    # += يعنى انه سيتم اضافه الحرف الى نهايه السلسله الموجوده في هذا العمود
    
    # تحويل الأعمدة إلى نص مشفر عن طريق دمج الأعمدة
    encrypted_text = ''.join(matrix)
    
    return encrypted_text

# دالة لفك تشفير النص باستخدام Columnar Transposition Cipher
def decrypt_message(ciphertext, key):
    # حساب عدد الأعمدة (طول المفتاح)
    num_columns = key
    
    # حساب عدد الصفوف
    num_rows = math.ceil(len(ciphertext) / num_columns)
    
    # حساب عدد الخانات الفارغة في الشبكة
    empty_cells = (num_columns * num_rows) - len(ciphertext)
    
    # ملء الأعمدة بالنص المشفر
    matrix = [''] * num_columns
    column = 0
    row = 0
    #جعلناهن =0 لكي ترجع الى العمود الاول بشكل صحيح
    
    for char in ciphertext:
        matrix[column] += char
        column += 1
        
        # عندما نملأ عمودًا، ننتقل إلى العمود التالي
        if column == num_columns or (column == num_columns - empty_cells and row >= num_rows - 1):
            column = 0
            row += 1
    
    #شرط اذا تم ملء جميع الاعمده او اذا وصلنا الى عمود يحتوي على عدد اقل من الحروف في هذه الحاله ننتقل الى العمود الاول والصف التالي
    
    
    # تحويل الشبكة إلى نص أصلي
    #سلسله فارغه لتخزين النص المفكوك
    decrypted_text = ''
    
    for r in range(num_rows):
        #تمر هذه الحلقه على الصفوف حيث r للصفوف
        for c in range(num_columns):
            #تمر هذه الحلقه على الاعمده 
            if len(matrix[c]) > r:
                #يتحقق اذا كان العمود يحتوي على حرف في الصف الحالي 
                decrypted_text += matrix[c][r]
              # يتم استعاده الحرف من الصف والعمود ويتم اضافته الى النص المفكوك 
    return decrypted_text

# المدخلات من المستخدم
plaintext = input("أدخل النص المراد تشفيره: ").replace(" ", "")
key = int(input("أدخل الارتفاع (المفتاح): "))

# تشفير النص
encrypted_text = encrypt_message(plaintext, key)
print("النص المشفر:", encrypted_text)

# فك التشفير
decrypted_text = decrypt_message(encrypted_text, key)
print("النص المفكوك:", decrypted_text)
```
